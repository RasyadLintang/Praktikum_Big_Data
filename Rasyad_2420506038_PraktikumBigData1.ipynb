# **Latihan 1**

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

# **Tugas 5**

```python
from pyspark.sql import SparkSession
import pandas as pd

# Spark Session
spark = SparkSession.builder.appName("BigDataPractice_5").getOrCreate()

data_pyspark = [("Acop", 17, "Mahasiswa", "Bermain Game", "Laki-Laki"), ("Asyrof", 18, "Mahasiswa", "Membuat Game", "Laki-Laki"), ("Asy", 19, "Mahasiswa", "Memperbaiki Game", "Laki-Laki"), ("Rof", 20, "Mahasiswa", "Merusak Game", "Laki-Laki")]
columns_pyspark = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_pyspark = spark.createDataFrame(data_pyspark, columns_pyspark)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pyspark_to_pandas = df_pyspark.toPandas()

# Data Pandas
data_pandas = {"Nama": ["Acop", "Asyrof", "Asy", "Rof"]}
df_data_pandas = pd.DataFrame(data_pandas)

# Gabungan Dataframe
df_merged = pd.merge(df_pyspark_to_pandas, df_data_pandas, on="Nama")
print("Gabungan dataframe:")
print(df_merged)

# Rata-rata usia
rata_rata_usia = df_merged['Usia'].mean()
print(f"\nRata-rata usia data gabungan: {rata_rata_usia}")
```
