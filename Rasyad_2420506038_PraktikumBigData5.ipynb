# **Tugas 1**

```python
data = "Rasyad adalah gold laner yang paling jago main mm di gold lane"

def map_function(data):
    for word in data.split():
        yield (word, 1)

print("Hasil Mapping: ")
list(map_function(data))
```

# **Tugas 2**

```python
def map_function(text):
    for word in text.split():
        yield (word, 1)
from collections import defaultdict

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return result
text = "ikan ayam ikan ayam sapi ikan"

mapped = list(map_function(text))
print("Hasil Map:", mapped)

reduced = reduce_function(mapped)
print("Hasil Reduce:", dict(reduced))
```
