```python
#Tugas1
from pymongo import MongoClient

# Inisialisasi client dan koneksi ke database
client = MongoClient('mongodb://localhost:27017/')
db = client['company_db']
collection = db['employees']

# Contoh untuk memastikan koneksi
print('Connected to MongoDB')
```

```python
#Tugas2
# Contoh Insert Data
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}

collection.insert_one(employee_data)
print("Data Inserted")

# Insert data sekaligus (multiple insert)
multiple_employees = [
    {'name': 'Ilham', 'department': 'Manager', 'age': 30, 'salary': 10000},
    {'name': 'Louis', 'department': 'IT', 'age': 25, 'salary': 5200},
    {'name': 'Deja', 'department': 'Finance', 'age': 24, 'salary': 6000},
    {'name': 'Sumbul', 'department': 'HR', 'age': 31, 'salary': 5800}
]

collection.insert_many(multiple_employees)
print("Data succesfully added!")
print("Employee Data")
for emp in collection.find():
    print(emp)
```

```python
# Update data
collection.update_many(
    {'department': 'IT'},
    {'$mul': {'salary': 1.15}}  # Naik 10%
)

print("Salary for department IT +15%")

# Menampilkan hasil
for emp in collection.find({'department': 'IT'}):
    print(emp)
```

```python
# Delete data
collection.delete_one({'name': 'Alice'})
print("Alice's data has been deleted!")
```

```python
# Menampilkan data setelah di update dan delete
print("Final Data")
for employee in collection.find():
    print(employee)
```

```python
#Tugas3
# Query Aggregation untuk mencari rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
#Latihan 4.1
from pymongo import MongoClient

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["rass_bunker"]
products = db["products"]

# Produk (Harga dalam bentuk $)
data_product = [
    {'name': 'Pistol Mk II', 'category': 'Weapon', 'price': 73750, 'stock': 20},
    {'name': 'SNS Pistol Mk II', 'category': 'Weapon', 'price': 79575, 'stock': 20},
    {'name': 'Heavy Revolver Mk II', 'category': 'Weapon', 'price': 99000, 'stock': 10},
    {'name': 'Pump Shotgun Mk II', 'category': 'Weapon', 'price': 82500, 'stock': 15},
    {'name': 'SMG Mk II', 'category': 'Weapon', 'price': 85500, 'stock': 10},
    {'name': 'Assault Rifle Mk II', 'category': 'Weapon', 'price': 98750, 'stock': 10},
    {'name': 'Carbine Rifle Mk II', 'category': 'Weapon', 'price': 107500, 'stock': 5},
    {'name': 'Bullpup Rifle Mk II', 'category': 'Weapon', 'price': 105750, 'stock': 5},
    {'name': 'Explosive Rounds/box', 'category': 'Ammo', 'price': 55000, 'stock': 20},
    {'name': 'AR Ammo/box', 'category': 'Ammo', 'price': 26500, 'stock': 50}
]

# Insert data ke koleksi
products.insert_many(data_product)
print("Data Products Inserted")
```

```python
print("Product List")
for product in products.find():
    print(product)
```

```python
#Latihan 4.2
# Rata-rata
avg_price = list(products.aggregate([
    {'$group': {'_id': None, 'avgPrice': {'$avg': '$price'}}}
]))[0]['avgPrice']

print(f"Rata-rata harga produk: {avg_price:.2f}\n")

# Menampilkan produk yang harganya di atas rata-rata
print("Produk dengan harga di atas rata-rata:")
for p in products.find({'price': {'$gt': avg_price}}):
    print(p)
```

```python
#Latihan 4.3
# Menghitung total produk setiap kategori
pipeline = [
    {'$group': {'_id': '$category', 'totalProduk': {'$sum': 1}}},
    {'$sort': {'totalProduk': -1}}
]

print("Total products in each category: ")
for result in products.aggregate(pipeline):
    print(result)
```

```python
#Tugas 5.1
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["perusahaan"]
karyawan = db["karyawan"]

data = [
    {"nama": "Ilham", "umur": 30, "gaji": 75000000, "departemen": "Manager"},
    {"nama": "Asep", "umur": 28, "gaji": 4500000, "departemen": "HR"},
    {"nama": "Nahmier", "umur": 29, "gaji": 5500000, "departemen": "HR"},
    {"nama": "Flip", "umur": 28, "gaji": 4000000, "departemen": "HR"},
    {"nama": "Jane", "umur": 27, "gaji": 6500000, "departemen": "Finance"},
    {"nama": "Kris", "umur": 28, "gaji": 5500000, "departemen": "Finance"},
    {"nama": "Hye Jin", "umur": 29, "gaji": 6000000, "departemen": "Finance"},
    {"nama": "Julio", "umur": 25, "gaji": 7200000, "departemen": "IT"},
    {"nama": "Ones", "umur": 28, "gaji": 7000000, "departemen": "IT"},
    {"nama": "Leren", "umur": 27, "gaji": 7100000, "departemen": "IT"}
]

# Masukkan data
karyawan.insert_many(data)
print("Data karyawan berhasil dimasukkan ke database 'perusahaan'!")
```

```python
pipeline_tugas1 = [
    # Pengelompokkan per departemen dan nama, ambil gaji tertinggi
    {"$group": {
        "_id": {"departemen": "$departemen", "nama": "$nama"},
        "gaji": {"$max": "$gaji"}
    }},
    
    # Pengurutan
    {"$sort": {"_id.departemen": 1, "gaji": -1}},
    {"$group": {
        "_id": "$_id.departemen",
        "top_5": {"$push": {"nama": "$_id.nama", "gaji": "$gaji"}}
    }},
    
    # 5 Karyawan Teratas
    {"$project": {"top_5": {"$slice": ["$top_5", 5]}}}
]

hasil = karyawan.aggregate(pipeline_tugas1)
for data in hasil:
    print(f"\nDepartemen: {data['_id']}")
    for emp in data['top_5']:
        print(f"{emp['nama']} - Rp{emp['gaji']:,.0f}".replace(",", "."))
```

```python
#Tugas 5.2
hapus = karyawan.delete_many({"umur": {"$lt": 26}})
print(f"{hapus.deleted_count} karyawan telah dihapus dari database.")

print("Data Setelah dihapus")
for emp in karyawan.find():
    print(emp)
```

```python
#Tugas 5.3
pipeline_tugas3 = [
    {"$group": {
        "_id": "$departemen",
        "total_gaji": {"$sum": "$gaji"},
        "rata_umur": {"$avg": "$umur"}
    }}
]

hasil3 = list(karyawan.aggregate(pipeline_tugas3))
for h in hasil3:
    print(f"Departemen: {h['_id']}")
    print(f"  Total Gaji: Rp{h['total_gaji']:,}")
    print(f"  Rata-rata Umur: {h['rata_umur']:.1f} tahun\n")
```
