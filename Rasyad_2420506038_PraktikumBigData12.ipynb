```python
#Latihan1
# Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df.fillna({
    'Age': df['Age'].mean(),
    'Salary': df['Salary'].median()
}, inplace=True)

df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
#Latihan2
# Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
#PracticeTask
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df = pd.read_csv('/home/rasyad/Downloads/Titanic - Titanic.csv')

print("Jumlah missing value per kolom:")
print(df.isna().sum())
```

```python
# Transformasi data untuk normalisasi kolom numerik
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Memperbaiki missing value numerik dengan rata-rata (mean)
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Melakukan normalisasi dengan MinMaxScaler
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\nNormalisasi kolom numerik:\n", df[num_cols].head())
```

```python
# Standarisasi kolom kategori
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Kolom kategori:", categorical_cols)

# Mengubah semua ke huruf kecil + kapitalisasi
for col in categorical_cols:
    df[col] = df[col].str.strip().str.lower().str.capitalize()

# Menghapus duplikat
df = df.drop_duplicates()

print("Result setelah standarisasi kategori:")
print(df[categorical_cols].head())
```
